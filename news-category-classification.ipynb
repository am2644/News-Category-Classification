```python
#loading dataset
df = pd.read_json("/kaggle/input/news-category-dataset/News_Category_Dataset_v3.json",lines=True)
df
```

```python
#cheking null cells in dataset
df.isna().sum()
```

```python
#checking number of records for each category
df['category'].value_counts()
```

the dataset is so big so I'm going to choose 5 category and for each category I'm chooding 5000 records 

```python
columns_lst = ['POLITICS','WELLNESS','ENTERTAINMENT','TRAVEL','STYLE & BEAUTY']
```

```python
#select the records with categories that listed in columns_list
df_t = df[df['category'].isin(columns_lst)]
df_t
```

```python
#select 5000 records for each categories
df_sample = df_t.groupby('category').head(5000)
df_sample
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

#transform categories to numeric values
df_sample_copy = df_sample.copy()
df_sample_copy['target'] = le.fit_transform(df_sample_copy['category'])
df_sample_copy.head()
```

```python
classes_dict = {le.classes_[i] : i for i in range(len(le.classes_))}
classes_dict
```

```python
#combine headline and short_description columns to new column(text)
df_sample_copy['text'] = df_sample_copy['headline'] +" " + df_sample_copy['short_description']
df_sample_copy['text'].iloc[20]
```

```python
from sklearn.model_selection import train_test_split

#split our dataset to text and train datasets
X_train, X_test, y_train, y_test = train_test_split(df_sample_copy['text'], df_sample_copy['target'], test_size=0.2,random_state=26)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
import spacy
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
#define a function for filter stop words and punctuations and extract lemma from the txts
def preprocesser(text_array):
    preprocessed_texts = []
    for text in text_array:
        doc = nlp(text)
        words_lst = []
        for token in doc:
            if not token.is_stop and not token.is_punct:
                words_lst.append(token.lemma_)
        preprocessed_text = " ".join(words_lst)
        preprocessed_texts.append(preprocessed_text)
    return preprocessed_texts
```

```python
text_processer = FunctionTransformer(preprocesser)
```

I'm going to train my model with MultinomialNB and LogisticRegression and vectorise the text by bag of wordas and n grams(1,3) algorithm using CountVectorizer

```python
cls1 = make_pipeline(
    text_processer,
    CountVectorizer(),
    MultinomialNB()
)



cls1.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls1.predict(X_test)))
```

```python
cls2 = make_pipeline(
    text_processer,
    CountVectorizer(ngram_range=(1,3)),
    MultinomialNB()
)



cls2.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls2.predict(X_test)))
```

```python
cls3 = make_pipeline(
    text_processer,
    CountVectorizer(),
    LogisticRegression(multi_class='multinomial', solver='lbfgs')
)



cls3.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls3.predict(X_test)))
```

```python
cls4 = make_pipeline(
    text_processer,
    CountVectorizer(ngram_range=(1,3)),
    LogisticRegression(multi_class='multinomial', solver='lbfgs')
)



cls4.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls4.predict(X_test)))
```

As we could see we get best result (%91 accuracy) from cls2 when we use  MultinomialNB and n_grams=(1,3)
